## Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
df = pd.read_csv('titanic_data.csv')
print df.describe()
```

```python
#print df.head()
# Replacing missing ages with median
print df['Age'].median()
df['Age'] = df['Age'].fillna(df['Age'].median())
df.loc[df["Survived"]==1, 'Survived'] = "Survived"
df.loc[df["Survived"]==0, 'Survived'] = "Died"
#df["ParentsAndChildren"] = df["Parch"]
#df["SiblingsAndSpouses"] = df["SibSp"]

plt.figure()
#sns.pairplot(data=df[['Fare', 'Survived', 'Pclass', 'Sex']], hue='Sex', dropna=True)
sns.pairplot(df, hue='Sex')
```
